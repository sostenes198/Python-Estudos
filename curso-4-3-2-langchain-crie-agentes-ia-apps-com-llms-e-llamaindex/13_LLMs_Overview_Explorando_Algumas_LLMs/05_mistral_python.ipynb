# 05 - Mistral com Python

**Mistral AI** é uma empresa francesa que desenvolve modelos open-source e proprietários.
Seus modelos são conhecidos pela eficiência — bom desempenho com menor custo computacional.

### Principais modelos

| Modelo | Uso recomendado |
|--------|-----------------|
| `mistral-large-latest` | Uso geral, alta qualidade |
| `mistral-small-latest` | Rápido e econômico |
| `codestral-latest` | Geração de código |
| `pixtral-12b-2409` | Análise de imagens (vision) |

> **`codestral-mamba-latest`** usado no original foi substituído por `codestral-latest` — versão atual.

### Pacotes necessários

```bash
pip install mistralai python-dotenv
```

> Crie sua API key em [console.mistral.ai](https://console.mistral.ai) e adicione `MISTRAL_API_KEY=sua_chave` no `.env`.

````python
import os
from dotenv import load_dotenv, find_dotenv
from mistralai import Mistral

load_dotenv(find_dotenv())

api_key = os.environ.get("MISTRAL_API_KEY")
client = Mistral(api_key=api_key)
````

## 1. Chat básico

A API do Mistral segue o padrão OpenAI — `client.chat.complete()` com lista de mensagens.

> **Atenção**: o método é `.complete()` (não `.create()` como na OpenAI).

````python
chat_response = client.chat.complete(
    model="mistral-large-latest",
    messages=[
        {
            "role": "user",
            "content": "Qual melhor jogador da Premier League?"
        }
    ]
)

print(chat_response.choices[0].message.content)
````

## 2. Vision — Análise de Imagens

O modelo `pixtral-12b-2409` aceita imagens via URL diretamente no campo `image_url`.
Diferente do Gemini (que exige base64), o Pixtral aceita URLs públicas diretamente.

````python
messages_vision = [
    {
        "role": "user",
        "content": [
            {
                "type": "text",
                "text": "Descreva para mim essa imagem"
            },
            {
                "type": "image_url",
                # Pixtral aceita URL diretamente — sem necessidade de base64
                "image_url": "https://tripfixers.com/wp-content/uploads/2019/11/eiffel-tower-with-snow.jpeg"
            }
        ]
    }
]

response_vision = client.chat.complete(
    model="pixtral-12b-2409",
    messages=messages_vision
)

print(response_vision.choices[0].message.content)
````

## 3. Geração de Código com Codestral

O **Codestral** é um modelo especializado em código, treinado em mais de 80 linguagens.
É significativamente melhor para tarefas de programação do que os modelos de propósito geral.

````python
response_code = client.chat.complete(
    model="codestral-latest",  # atualizado do codestral-mamba-latest
    messages=[
        {
            "role": "user",
            "content": "Escreva uma função Python para fibonacci com memoização"
        }
    ]
)

print(response_code.choices[0].message.content)
````

## Resumo

| Funcionalidade | Modelo | Método |
|---------------|--------|--------|
| Chat geral | `mistral-large-latest` | `client.chat.complete()` |
| Análise de imagem | `pixtral-12b-2409` | `content` com `image_url` |
| Geração de código | `codestral-latest` | `client.chat.complete()` |
