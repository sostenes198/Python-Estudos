# 06 - Mistral com LangChain

O `ChatMistralAI` integra os modelos Mistral ao ecossistema LangChain.
Interface idêntica ao `ChatOpenAI` e `ChatGroq`.

### Pacotes necessários

```bash
pip install langchain-mistralai python-dotenv
```

````python
from dotenv import load_dotenv, find_dotenv
from langchain_mistralai import ChatMistralAI

load_dotenv(find_dotenv())

# Sem especificar model, usa mistral-small por padrão
# Para maior qualidade: model="mistral-large-latest"
chat = ChatMistralAI(model="mistral-large-latest")

response = chat.invoke("Olá, você veio de onde?")
print(response.content)
````

## 1. Chat com mensagens de sistema

````python
mensagens = [
    ("system", "Você é especialista em traduzir do português para o alemão. Traduza a frase:"),
    ("human", "Hoje é um bom dia para estudar")
]

response = chat.invoke(mensagens)
print(response.content)
````

## 2. Chain com ChatPromptTemplate

O `ChatMistralAI` funciona perfeitamente com o operador `|` do LCEL.

````python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

template = ChatPromptTemplate.from_messages([
    ("system", "Você é um contador de anedotas em português"),
    ("human", "Crie uma história sobre: {anedota}")
])

chain = template | chat | StrOutputParser()

response = chain.invoke({"anedota": "formiga e cigarra"})
print(response)
````

## 3. Streaming com LCEL

````python
for chunk in chain.stream({"anedota": "leão e o rato"}):
    print(chunk, end="", flush=True)
````

## Resumo

| Operação | Código |
|----------|--------|
| Criar modelo | `ChatMistralAI(model="mistral-large-latest")` |
| Chat direto | `chat.invoke([("system", ...), ("human", ...)])` |
| Chain LCEL | `template \| chat \| StrOutputParser()` |
| Streaming | `chain.stream({...})` |

> **Vantagem do Mistral**: modelos open-source disponíveis para auto-hospedagem (via Ollama)
> e excelente custo-benefício nos modelos cloud.
