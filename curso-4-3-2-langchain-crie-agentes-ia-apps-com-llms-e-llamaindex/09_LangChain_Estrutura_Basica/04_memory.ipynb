```python
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()

# langchain.debug = false

chat = ChatOpenAI(model="gpt-3.5-turbo-0125")
```

```python
# Criando o histórico
history = InMemoryChatMessageHistory()

# Encapsulando o modelo com histórico
chain = RunnableWithMessageHistory(chat, lambda session_id: history)
```

```python
config = {"configurable": {"session_id": "minha-sessao"}}

def conversar(mensagem: str):
    resposta = chain.invoke(
        [HumanMessage(content=mensagem)],
        config=config
    )
    return resposta.content

# Conversa com memória
print(conversar("Meu nome é Sóstenes."))
print(conversar("Qual é o meu nome?"))
print(conversar("Me dê uma dica de Python."))
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.messages import HumanMessage
from dotenv import load_dotenv

load_dotenv()

chat = ChatOpenAI()

history = InMemoryChatMessageHistory()
chain = RunnableWithMessageHistory(chat, lambda session_id: history)

config = {"configurable": {"session_id": "minha-sessao"}}

def predict(input: str):
    resposta = chain.invoke(
        [HumanMessage(content=input)],
        config=config
    )
    return resposta.content

# Simulando o que o curso faz
print(predict("Faça uma poesia de 400 palavras"))
print(predict("Resuma a poesia anterior"))
```

```python
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.messages import trim_messages

history = InMemoryChatMessageHistory()

# Trimma o histórico para no máximo 250 tokens antes de enviar
chain = RunnableWithMessageHistory(
    trim_messages(max_tokens=250, strategy="last", token_counter=chat) | chat,
    lambda session_id: history
)
```
