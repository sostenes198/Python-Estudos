```python
from ultralytics import YOLO
```

# Detecção de objetos com YOLO

```python
# Exemplo
# Load a pretrained YOLO model (e.g., YOLOv8s for small version)
# Models are automatically downloaded on first use
model = YOLO("yolo11m.pt")

# Perform object detection on an image or video
# This example uses a sample image from the Ultralytics website
results = model("https://ultralytics.com/images/bus.jpg")

# Process and visualize results
for result in results:
    print(result.boxes)  # Print bounding box predictions
    result.show()        # Display the image with predictions
```
