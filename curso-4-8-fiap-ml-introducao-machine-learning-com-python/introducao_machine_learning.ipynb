```python
import numpy as np
```

# 1 - Importação e Criação de Arrays

```python
# Criando um array a partir de uma lista
lista = [1, 2, 3, 4, 5]
array1 = np.array(lista)
print(f'Array1: {array1} \n')

# Criando um array 2D (matriz)
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print(f'Matriz: {matriz} \n')
```

# 2 - Operações Básicas com Arrays

```python
# Soma, multiplicação e operações element-wise
array2 = np.array([10, 20, 30, 40, 50])
soma = array1 + array2
multiplicacao = array1 * array2

print(f'Soma: {soma} \n')
print(f'Multiplicacao: {multiplicacao} \n')
```

# 3 - Funções Numpy Comuns

```python
media = np.mean(array2)
desvio = np.std(array2)

print(f'Média: {media} \n')
print(f'Desvio: {desvio} \n')
```

# Pandas - Manipulação e Análise de Dados

Objetivo: Utilizar o Pandas para carregar, explorar e manipular conjuntos de dados de forma eficiente

# 1 - Importação e Criação de DataFrame

```python
import pandas as pd
```

```python
# Criando um DataFrame a partir de um dicionário
dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos'],
    'Idade': [25, 30, 35],
    'Salario': [50000, 80000, 100000]
}

df = pd.DataFrame(dados)
print(f'DataFrame: {df} \n')
```

# 2 - Manipulação e Exploração de Dados:

```python
# Selecionando uma coluna
print(f'Coluna "Nome": {df["Nome"]} \n')

# Filtrando linhadas com base em uma condição
df_filtrado = df[df['Salario'] > 60000]
print(f'Filtrado: {df_filtrado} \n')

# Estatísticas descritivas
print(f'Estatísticas: {df.describe()} \n')
```

# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# Gráfico de linha
plt.plot(df['Idade'], df['Salario'], marker='o')
plt.title('Salário x Idade')
plt.xlabel('Idade')
plt.ylabel('Salario')
plt.show()
```

# Seaborn

```python
import seaborn as sns

# Configuração padrão de gráficos mais bonitos
sns.set_theme(style='whitegrid')
```

```python
# Gráfico de dispersão (scatter plot)
sns.scatterplot(x='Idade', y='Salario', hue='Nome', data=df)
plt.title('Scatter plot: Salário vs Idade')
plt.show()
```

```python
# Boxplot para visualizar a distribuição do salário
sns.boxplot(y='Salario', data=df)
plt.title('Distribuição de Salário')
plt.show()
```

```python
# Gráfico de distriuição (histograma com KDE)
sns.histplot(df['Salario'], kde=True)
plt.title('Histrograma do Salário')
plt.show()
```
