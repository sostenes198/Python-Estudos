```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
```

```python
df = pd.read_csv('./kc_final.csv')
df.head()
```

```python
df.shape
```

```python
df = df[['id', 'date', 'price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront']]
df.head()
```

```python
df.shape
```

```python
colunas_para_remover = ['id', 'date']
df = df.drop(colunas_para_remover, axis=1)
```

```python
df.shape
```

```python
x = df.drop('price', axis=1)
y = df['price']
```

```python
min_max_scaler = StandardScaler()
x = min_max_scaler.fit_transform(x)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=23)
```

```python
print(f'Exemplos para treinamento: {len(y_train)}. E para o teste: {len(y_test)}')
```

# Regressão Linear

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test, y_pred)

print(f'MSE: {mse}\nRMSE: {rmse}\nMAPE: {mape}')
```

# K-NN Regressor

```python
model = KNeighborsRegressor(n_neighbors=7, metric='euclidean')
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test, y_pred)

print(f'MSE: {mse}\nRMSE: {rmse}\nMAPE: {mape}')
```

# SVR Support Vector Machines para Regressão

```python
model = SVR()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test, y_pred)

print(f'MSE: {mse}\nRMSE: {rmse}\nMAPE: {mape}')
```

```python
model = SVR(kernel='linear', C=100)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test, y_pred)

print(f'MSE: {mse}\nRMSE: {rmse}\nMAPE: {mape}')
```

# Árvore de Regressão

```python
model = DecisionTreeRegressor()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test, y_pred)

print(f'MSE: {mse}\nRMSE: {rmse}\nMAPE: {mape}')
```

# XGBoost

```python
model = XGBRegressor()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test, y_pred)

print(f'MSE: {mse}\nRMSE: {rmse}\nMAPE: {mape}')
```

```python
params = {"n_estimators": 100,
          "max_depth": 6,
          "learning_rate": 0.1}
```

```python
model = XGBRegressor(**params)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test, y_pred)

print(f'MSE: {mse}\nRMSE: {rmse}\nMAPE: {mape}')
```
