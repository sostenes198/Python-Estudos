```python
iris = datasets.load_iris()
```

```python
iris
```

```python
# O dataset Iris consiste em dados numéricos (medidas), não imagens.
# Para acessar a primeira amostra (primeira linha de dados):
primeira_amostra = iris.data[0]
print("Features da primeira amostra:", primeira_amostra)
print("Label da primeira amostra:", iris.target[0])
print("Nome da classe:", iris.target_names[iris.target[0]])
```

```python
# Visualizando as features da primeira amostra com matplotlib
plt.figure(figsize=(8, 6))
plt.bar(iris.feature_names, iris.data[0])
plt.title('Features da primeira amostra do dataset Iris')
plt.ylabel('Valor (cm)')
plt.show()
```
